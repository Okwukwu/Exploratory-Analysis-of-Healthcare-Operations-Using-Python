There are 10,000 patients in the dataset with 15 columns which contain the patient's details such as name, age, gender, blood group, medical conditions and other details.
```python
#check for null in the admission and discharge column
df[['Date of Admission', 'Discharge Date']].isna().sum() 
```

```python
# check for invalid dates where discharge date comes before admission date
invalid_dates = df[df['Discharge Date'] < df['Date of Admission']]

invalid_dates
```

```python
# feature engineering
#creating length of stay column

df['length_of_stay'] = (df['Discharge Date'] - df['Date of Admission']).dt.days 

df['length_of_stay']
```

```python
#check for negative (invalid) length of stay

df[df['length_of_stay'] < 0]
```

```python
#statistics of length of stay

df['length_of_stay'].describe()

#statistics of billing amount
df['Billing Amount'].describe()
```

```python
# checking if certain medical conditions always involve specific medications

single_med_conditions = df.groupby('Medical Condition')['Medication'].nunique()
single_med_conditions = single_med_conditions[single_med_conditions == 1]
single_med_conditions
```

```python
# Top 10 doctors that handled the widest variety of conditions
doc_cond_count = df.groupby('Doctor')['Medical Condition'].nunique().sort_values(ascending=False)

doc_cond_count.head(10)
```

```python
# The doctor handling the widest variety of conditions and the number of condition

max_cond = doc_cond_count.max()

doc_cond_count[
    doc_cond_count == max_cond
    ]
```

```python
#Conditions treated by the doctors that attended to wide variety of cases

top_doc = doc_cond_count[doc_cond_count == max_cond].index

df[df['Doctor'].isin(top_doc)][['Doctor', 'Medical Condition']].drop_duplicates()
```

```python
#average length of stay per doctor

max_stay_doc = df.groupby('Doctor')['length_of_stay'].mean().sort_values(ascending=False)
max_stay_doc.head(10)
```

```python
#proportion normal vs abnormal test results

df['Test Results'].value_counts()
```

```python
#proportion normal vs abnormal test results in percentage

df['Test Results'].value_counts(normalize=True) * 100
```

```python
#checking if abnormal test results are associated with longer hospital stay

df.groupby('Test Results')['length_of_stay'].agg(['mean', 'median', 'count'])
```

```python
#medical conditions with highest abnormal test results rate

df['abnormal_result'] = df['Test Results'] == 'Abnormal'

Abnormal_result = (
    df.groupby('Medical Condition')['abnormal_result'].count().sort_values(ascending=False)
)
Abnormal_result
#df.groupby('Medical Condition')[['Test Results'] == 'Abnormal'].mean()
```
